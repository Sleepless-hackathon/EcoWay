```python
class Aqi():
    VERY_GOOD = 33
    GOOD = 66
    FAIR = 99
    POOR = 149
    VERY_POOR = 200
    HAZARDOUS = 10000
    
    values = {
        VERY_GOOD: 10,
        GOOD: 8,
        FAIR: 6,
        POOR: 4,
        VERY_POOR: 2,
        HAZARDOUS: 0
    }
```

```python
aqi = sample_df["AQI"]
```

```python
air_score = []

for a in aqi:
    if a <= Aqi.VERY_GOOD:
        air_score.append(Aqi.values[Aqi.VERY_GOOD])
    elif a <= Aqi.GOOD:
        air_score.append(Aqi.values[Aqi.GOOD])
    elif a <= Aqi.FAIR:
        air_score.append(Aqi.values[Aqi.FAIR])
    elif a <= Aqi.POOR:
        air_score.append(Aqi.values[Aqi.POOR])
    elif a <= Aqi.VERY_POOR:
        air_score.append(Aqi.values[Aqi.VERY_POOR])
    else:
        air_score.append(Aqi.values[Aqi.HAZARDOUS])
```

```python
sample_df["air_score"] = air_score
```

```python
import random, math
```

```python
random.seed(42)
```

```python
def point_in_radius(o_lat, o_lon, r_meters):
    r = r_meters / 111300
    u = random.random()
    v = random.random()
    w = r * math.sqrt(u)
    t = 2 * math.pi * v
    x = w * math.cos(t)
    lat_shift = w * math.sin(t)
    lon_shift = x / math.cos(lat_shift)
    
    return o_lat + lat_shift, o_lon + lon_shift
```

```python
center = (0, 0)
```

```python
points = []

for _ in range(1000):
    points.append(point_in_radius(center[0], center[1], 10000))
```
