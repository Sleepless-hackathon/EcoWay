```python
import osmap_utils
```

```python
import pandas as pd
```

```python
dataset = pd.read_pickle("./names_aqi_median.pkl")
```

```python
dataset
```

```python
for c in dataset:
    col = dataset[c]
    print(col.dropna())
```

```python
cord_to_value = {}
```

```python
finder = osmap_utils.StreetSearcher(city="Москва")
```

```python
import matplotlib.pyplot as plt
```

```python
for c in dataset.keys():
    sensors = list(dataset[c].dropna())
    cords = [ct.get_cord() for ct in finder.query_cords(c)]
    
    for si in range(len(sensors)):
        cord_to_value[cords[si % len(cords)]] = sensors[si]
        
```

```python
cord_to_value
```

```python
len(cord_to_value)
```

```python
X = [k for k in cord_to_value.keys()]
Y = [v for v in cord_to_value.values()]
```

```python
X
```

```python
Y
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(X)
X_copy = X

X = scalar.transform(X)
```

```python
model.add(Dense(16, input_dim=2, activation="relu"))
```

```python
model.add(Dense(8, activation="linear"))
```

```python
model.add(Dense(4, activation="linear"))
```

```python
model.add(Dense(1, activation="linear"))
```

```python
import tensorflow as tf
```

```python
optimizer = tf.keras.optimizers.RMSprop(0.0099)
```

```python
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(X,Y,epochs=500)
```

```python
model.predict([(55.7467696, 37.5727272), (55.7288191, 37.547184), (55.6587058, 37.6410313)])
```

```python
model.save("./aqi.model")
```
