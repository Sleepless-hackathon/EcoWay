```python
import pandas as pd
import os
dataset = "./data-csv/"
if not os.path.exists(dataset):
    os.system("./download_data.sh")
files = os.listdir(dataset)
files.sort()
```

```python
dataframes = []
```

```python
files
```

```python
for file in files:
    full_path = os.path.join(dataset, file)

    name, tail = file.split("_")
    sensor_n, tail = tail.split(".")
    name = name.strip()
    sensor_n = int(sensor_n)

    print(name, "|", sensor_n)

    df = pd.read_csv(full_path)
    dataframes.append({"name": name, "number": sensor_n, "df": df})
```

```python
AQI_by_names = {}

# print(AQI_by_cords)

for part in dataframes:
    if part["name"] not in AQI_by_names.keys():
        AQI_by_names[part["name"]] = {}
    
    AQI = part["df"]["AQI"]
    print(AQI)
    AQI_by_names[part["name"]][part["number"]] = AQI.mean()
```
